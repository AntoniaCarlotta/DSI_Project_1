##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
def my_avg(x):
    avg = sum(x)*1.0/len(x)
    return avg
```

```python
my_avg([2,3])
```

```python
def stdev(x):
    avg = my_avg(x)
    sdev = (sum([(i - avg)**2 for i in x])/len(x))**0.5
    return sdev

print stdev(rate)
print stdev(math)
print stdev(verbal)
```
